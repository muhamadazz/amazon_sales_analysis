**DATA CLEANING AMAZON SALES**

```python
# --- Konversi Kolom Harga dan Rating ke Numerik ---
def clean_price(x):
    return float(str(x).replace("‚Çπ", "").replace(",", "").strip()) if pd.notnull(x) else np.nan

def clean_percentage(x):
    return float(str(x).replace("%", "").strip()) if pd.notnull(x) else np.nan

def clean_rating_count(x):
    return int(str(x).replace(",", "").strip()) if pd.notnull(x) else np.nan

df['discounted_price'] = df['discounted_price'].apply(clean_price)
df['actual_price'] = df['actual_price'].apply(clean_price)
df['discount_percentage'] = df['discount_percentage'].apply(clean_percentage)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = df['rating_count'].apply(clean_rating_count)
df.head()
```

```python
# --- Imputasi Nilai Hilang ---
df['rating_count'].fillna(df['rating_count'].median(), inplace=True)
df['rating'].fillna(df['rating'].median(), inplace=True)

# --- Cek Kembali Missing Values ---
print("\n‚úÖ MISSING VALUES (SETELAH PEMBERSIHAN)")
print(df.isnull().sum())
```

```python
# --- Statistik Deskriptif setelah Dibersihkan ---
print("\nüìä STATISTIK DESKRIPTIF (Setelah Dibersihkan)")
df[['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']].describe()
```

üìä **Exploratory Data Analysis (EDA)**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker

# Style
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
#Distribusi Rating Produk
sns.histplot(df['rating'], bins=20, kde=True, color='skyblue')
plt.title('Distribusi Rating Produk')
plt.xlabel('Rating')
plt.ylabel('Jumlah Produk')
plt.show()
```

```python
#Distribusi Harga dan Diskon
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.histplot(df['actual_price'], bins=30, kde=True, ax=axes[0], color='salmon')
axes[0].set_title('Distribusi Harga Asli')
axes[0].set_xlabel('Harga (‚Çπ)')

sns.histplot(df['discount_percentage'], bins=30, kde=True, ax=axes[1], color='lightgreen')
axes[1].set_title('Distribusi Diskon (%)')
axes[1].set_xlabel('Diskon (%)')

plt.tight_layout()
plt.show()
```

```python
#3. Korelasi antar variable
sns.heatmap(df[['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']].corr(),
            annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Korelasi Antar Variabel Numerik')
plt.show()
```

```python
#10 Kategori dengn produk terbanyak
top_categories = df['category'].value_counts().head(10)

sns.barplot(x=top_categories.values, y=top_categories.index, palette='viridis')
plt.title('10 Kategori Produk Terbanyak')
plt.xlabel('Jumlah Produk')
plt.ylabel('Kategori')
plt.show()
```
